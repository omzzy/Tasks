Creating a class in order quickly create graphs and tables for each year for the gross claims incurred, from here any outliers or interesting firms will be looked at in greater detail

```python
class Monitor:
    def __init__(self,df,column,year):
        self.df = df
        self.column = column
        self.year = year
    
    def graphOfallValues(self):
        print(self.df[self.column][self.year].describe())
        return self.df[self.column][self.year].plot(kind="hist",bins=80,title=self.year+" "+self.column)
    
    def percentOverTime(self):
        print(self.df[self.column].replace(to_replace=0,value=np.nan).pct_change(axis=1)[self.year].describe())
        return self.df[self.column].replace(to_replace=0,value=np.nan).pct_change(axis=1)[self.year].plot(kind="hist",bins=80,title="Percentage change"+" "+self.year+" "+self.column)
    
    def percentData(self):
        return self.df[self.column].replace(to_replace=0,value=np.nan).pct_change(axis=1)[self.year]
```

```python
# Combining both dataframes in order to check the SCR coverage ratio and Net combined ratio, these give a good
# indication of a firms capacity to pay claimants as well as the ratio they have paid out for claimants for that
# particular year

df_combined = df11.join(df22,how="inner")
df_combined
```

These attributes are important as if the SCR coverage ratio is less than 1 and the Net combined ratio is more than 1, the firm would not have enough funds to pay off the claimants

```python
def coverageCombined(year):
    scrNetcombined = df_combined[["SCR coverage ratio","Net combined ratio"]].xs(key=year,axis=1,level=1)
    scrNetcombined = scrNetcombined[(scrNetcombined["SCR coverage ratio"] < 1) & (scrNetcombined["Net combined ratio"] > 1)]
    print(sns.jointplot(x=scrNetcombined['SCR coverage ratio'],y=scrNetcombined['Net combined ratio']))
    return scrNetcombined
```

## Deeper insight with machine learning models

```python
df_combined
```

```python
# Importing the StandardScaler function in order to scale the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = df_combined
```

```python
# Scales all values within the dataset to a number between 0-1

X = scaler.fit_transform(X)
X
```

```python
# Using the KMeans clustering machine learning algorithm as this is an unsupervised technique

from sklearn.cluster import KMeans
```

```python
# Creating a model that will produce 2 clusters from the data points

kmeans = KMeans(n_clusters=2)
```

```python
# Fitting the dataset to the model

kmeans.fit(X)
```

```python
# Getting the cluster centre vectors

kmeans.cluster_centers_
```

```python
# Sum of squared distances of samples to their closest cluster center, weighted by the sample weights if provided.

kmeans.inertia_
```

```python
# The predictions made by the model:
kmeans.labels_
```

```python
#Creaing a series in order to get 2 levels so that I can join these figures to the original dataset

index = pd.MultiIndex.from_product([["Predictions"],df_combined.index])
```

```python
dfKmean = pd.Series(data=kmeans.labels_,index=index).to_frame().unstack(0)
dfKmean
```

```python
dfKmean.rename(columns={0:"Predictions","Predictions":"2020YE"},inplace=True)
dfKmean
```

```python
# Joined the KMeans figures to dataset

df_combined1 = df_combined.join(dfKmean,how="inner")
df_combined1
```

```python
df_combined1["Predictions"]["2020YE"][df_combined1["Predictions"]["2020YE"] == 0]
```
